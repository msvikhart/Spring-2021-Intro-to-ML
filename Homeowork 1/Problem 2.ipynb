```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Folds5x2_pp.csv", sep = ',').to_numpy()
cols = ['Temperature', 'AmbientPressure','RelativeHumidity', 'ExhaustVacuum', 'ElectricalEnergy']
df = pd.DataFrame(data , columns = cols)
df
```

```python
X = df.drop(columns = 'ElectricalEnergy').to_numpy()
y = df['ElectricalEnergy'].to_numpy()

(TrainX, testX, Trainy, testy) = train_test_split(X, y, test_size = .25, random_state = 1)

model = LinearRegression(fit_intercept = False)
model.fit(TrainX, Trainy)

predictions = model.predict(testX)

parameters = model.coef_

print('The r^2 score is', r2_score(testy, predictions))
print('The mean absolute error is', mean_absolute_error(testy, predictions))

```

```python
def univariatePolynomialBasis(M):
    def polynomialM(x):
        
        out = np.array([])
        
        for i in range(M+1):
            out = np.append(out, x ** i)
              
        return out
    return polynomialM
        
```

```python

poly = univariatePolynomialBasis(2)

Xh = np.apply_along_axis(poly,1,X)

(TrainXh, testXh, Trainy, testy) = train_test_split(Xh, y, test_size = .25, random_state = 1)

model = LinearRegression(fit_intercept = False)
model.fit(TrainXh, Trainy)

predictions = model.predict(testXh)

parameters = model.coef_


print('The r^2 score is', r2_score(testy, predictions))
print('The mean absolute error is', mean_absolute_error(testy, predictions))
```
